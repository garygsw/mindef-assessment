##  Get distance to nearest expressway

```python
# https://beta.data.gov.sg/datasets/d_29f066d67df3eae91df8a42f443863c8/view
filepath = "../data/raw/NationalMapPolygonKML.geojson"
sg = gpd.read_file(filepath)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sg.plot(ax=ax, alpha=0.5, color="black")
expressways_df.plot(color="red", ax=ax)
df.plot.scatter(x="lon", y="lat", color="orange", ax=ax)
```

```python
A = df[["lon", "lat"]].values
B = [np.array(geom.coords)[:,:-1] for geom in expressways_df.geometry.to_list()]
B_ix = tuple(itertools.chain.from_iterable(
    [itertools.repeat(i, x) for i, x in enumerate(list(map(len, B)))]
))
B = np.concatenate(B)
ckd_tree = cKDTree(B)
dist, idx = ckd_tree.query(A, k=1)
idx = itemgetter(*idx)(B_ix)
dataset = pd.concat([
    df.reset_index(drop=True),
    expressways_df.loc[idx, ["name"]].reset_index(drop=True),
    pd.Series(dist, name='dist')],
    axis=1
)
```

## Find correlation between resale prices and distance to nearest expressway

```python
dataset["storey_range"].value_counts().sort_index()
```

```python
dataset["floor"] = dataset["storey_range"].astype(str).str[:2].astype(int)
```

```python
dataset["price_psm"] = dataset["resale_price"] / dataset["floor_area_sqm"]
```

```python
sns.lmplot(
    data=dataset,
    x="dist",
    y="price_psm",
    hue="floor",
    fit_reg=False,
    palette="RdYlGn_r",
    aspect=2.5
) 
```

```python
# overall pearson correlation
dataset["price_psm"].corr(dataset["dist"])
```

```python
storey_range_corr = dataset.groupby("storey_range")[["price_psm", "dist"]].corr().unstack().iloc[:,1]
storey_range_corr.name = "storey_range_corr"
storey_range_corr
```

```python
sns.heatmap(pd.DataFrame(storey_range_corr), annot=True, vmin=-1, vmax=1, cmap="RdYlGn")
```
